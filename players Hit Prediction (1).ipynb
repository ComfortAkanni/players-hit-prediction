**Build a prediction model to predict the number of Hits of each payer**

```python
#machine learning algorithms will e used to build a prediction model for the number of hits of each player .
#since we want to predict a continuous target variable,regression models will e used.
```

```python
#importing important lirbrary

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
```

To build a prediction model, we need to select the input features (independent variables) and the target variable (dependent variable) for the prediction. In this case, the target variable is the number of hits ('Hits') of each player, and the input features are the columns that can be used to predict the number of hits.

```python
#create a new data frame for the model
# Select the input features and target variable
selected_columns = [
     'PlayerName', 'Age', 'Preferred Foot', 'W/F1', 'SM1', 'IR1',  'value1', 'wage1', 'releaseClause1',  'wt_kg', 'height',
      'Height_Category', 'play position LW', 'play position LM', 'play position CAM', 'play position RW', 'play position CF', 'play position CB', 'play position RWB',
     'play position RB', 'play position RM', 'play position CM', 'play position LB', 'play position LWB', 'play position ST', 'play position CDM', 'play position GK'
]
target_column = 'Hits'

# Create the new DataFrame with only the selected columns
dfModel = dfNew[selected_columns + [target_column]]
```

```python
dfModel
```

```python
# Filter numerical columns
numerical_columns = dfModel.select_dtypes(include=['int64', 'float64']).columns

# Use only numerical columns for building the model
X = dfModel[numerical_columns].drop(columns=['Hits'])
y = dfModel['Hits']

# Impute missing values in X
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Create and train the Random Forest Regressor model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
